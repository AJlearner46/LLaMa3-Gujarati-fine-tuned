```python
# Step 3: Create Repository and Upload Model
from huggingface_hub import HfApi
from huggingface_hub import upload_folder
```

```python
# Define your model path and repository details
model_path = '/content/drive/MyDrive/Llama3/model-unsloth.Q4_K_M.gguf'
repo_name = 'llama3_gujarati_4bit_gguf'
user_name = 'ayushrupapara'
```

```python
# Create a HfApi instance
api = HfApi()
```

```python
# Upload the model file to the repository
api.upload_file(
    path_or_fileobj=model_path,
    path_in_repo='model.gguf',
    repo_id=f'{user_name}/{repo_name}',
    repo_type='model'
)
```
